Python Variables
</h2>

</div>

#### **What is a Variable?**
A variable in Python is like a labeled container used to store data. When you assign a value to a variable, Python creates an object in memory and attaches the variable name to it.

- Python is dynamically typed, so you dont need to declare the type of the variable.
- A variable gets its type from the value assigned to it.

#### **Creating a Variable**
You assign a value using the `=` operator.

```python
x = 10            # Integer
name = "Alice"    # String
price = 99.99     # Float
is_active = True  # Boolean

print(x)
print(name)
print(price)
print(is_active)
```

**Behind the scenes, Python does three things:**
- Creates an object with the value (`10`, `"Alice"`, etc.)
- Assigns that object a type
- Links the variable name (`x`, `name`) to the object

#### **Python Variable Names**
Rules for Naming Variables:
- Variable names can only contain letters (AZ, az), digits (09), and underscores (_)
- Must start with a letter or underscore (`_var`, `name`, but not `1value`)
- Cannot start with a digit (`2name` )
- Cannot use Python keywords (`for`, `if`, `class`, `def`, etc.)
- Case-sensitive: `Name` and `name` are different variables

#### **Best Practices for Variable Names**

- Use **descriptive names**

```python
n = 90          # bad practice
student_age = 90  # good practice
```

- Use **snake_case** for multiple words (standard in Python):

```python
max_score = 100
user_name = "rishabh"
```

#### **Invalid Examples**

```python
1st_place = "Gold"     #  starts with number
my-name = "Alice"      #  hyphen not allowed
for = 5                #  'for' is a keyword
```

#### **Valid Examples**

```python
first_place = "Gold"
my_name = "Alice"
_name123 = "Hidden"
```

#### **Check If a Name Is a Keyword**
**Keywords** are **reserved words** in Python that have **special meanings**. They are part of the syntax of the language and are used to define the structure and flow of Python code.

You cannot use keywords as variable names, function names, or identifiers because Python uses them to understand and interpret your code.
Some of these keywords are: `if`, `else`, `elif`, `while`, `for`, `break`, `continue`, `and`, `or`, `not`, `is`, `in`, etc.

```python
import keyword
print(keyword.iskeyword("for"))    # True
print(keyword.iskeyword("name"))   # False
```

Python Data Types
</h2>

</div>

Python supports various types: integers, floats, strings, lists, tuples, dictionaries, etc.

**Why Are Data Types Important?**
- Memory efficiency: Python stores different data types differently in memory.

- Operations: Each type supports different methods and operations.

- Error prevention: Knowing data types helps avoid bugs caused by wrong assumptions (e.g., trying to add a string to a number).

- Type checking: Python is dynamically typed, but still does type checking behind the scenes.

Categories are -->

#### **Numbers**
These include integers and floating-point numbers.

- **Integers (`int`)**: Whole numbers

Examples: `5`, `-3`, `42`

- **Floats (`float`)**: Numbers with a decimal point

Examples: `3.14`, `-0.001`

**Uses:**
- `int:` Counting, indexing, discrete quantities.

- `float:` Measurements, percentages, scientific values.

---

#### **Strings (`str`)**
A string is an ordered, immutable sequence of characters. Strings support indexing, slicing, and many built-in methods.

- **Single-line String:** - `"Hello"`
- **Multi-line String:** - `"""Hello, World!"""`

**Uses:** Text processing, messages, file names, web content, etc.

---

#### **Booleans (`bool`)**
A `bool` represents one of two values:

- `True`

- `False`

**Uses:** Conditional statements, logical control flow.

---

#### **Lists**
Ordered, **mutable** collections of items, which can be of mixed types.

**Uses:** Collection of items, arrays, storing dynamic sequences.

Example: [`1`, `"apple"`, `3.14`, `True`]

---

#### **Tuples**
A tuple is like a list, but immutable. 
It is a ordered, immutable collections of items.

**Uses:** Fixed collections, function returns, coordinates, dictionary keys.

Example: (`3`, `"apple"`, `8.5`)

---

#### **Dictionaries**
A dictionary is an unordered (ordered as of Python 3.7+) collection of `key: value` pairs.

**Uses:** When data needs to be accessed by name or label.

Example: {"name": "Alice", "age": 25}

---

#### **Sets**
A set is unordered, unindexed, no duplicate elements

**Uses:** Removing duplicates, performing set operations (union, intersection)

Example: `{4, 5, 6, 6}`  `{4, 5, 6}`

---

#### **None**
`None` is a special constant representing **null** or **no value**. Represents the absence of a value or a null value.

Example: x = None

---

Each data type serves different purposes and has different methods associated with it. For example, lists and dictionaries have methods to modify their content, while tuples and strings do not.

Python Casting
</h2>

</div>

In Python, every value has a data type, like integer (`int`), text (`str`), or a number with a decimal (`float`).

Sometimes, you get data in one type but need it in another. For example, when you scrape a price from a website, it often comes as a string (e.g., `"29.99"`), but you need it as a number to do calculations.

Type casting is the process of deliberately converting a variable from one data type to another. In Python, you do this using special constructor functions.

### Key Casting Functions:
`int()`: Converts a value into an integer (a whole number). It truncates (chops off) decimals, it does not round.



```python
# From a float
price_as_integer = int(29.99)  

# From a string
product_id = int("123")        

print(price_as_integer)
print(product_id)
print(type(price_as_integer))
print(type(product_id))
```

`float()`: Converts a value into a floating-point number (a number with a decimal).

```python
# From an integer
item_count = float(10)        

# From a string
price_as_float = float("29.99") 

print(item_count)
print(price_as_float)
print(type(item_count))
print(type(price_as_float))
```

`str()`: Converts a value into a string (text). This is useful for combining numbers with text.

```python
# From an integer
product_id_str = str(123)

# From a float
price_str = str(29.99)

# Example of use:
print("The price is $" + price_str) 
```

Python Operators
</h2>

</div>

Operators are used to perform operations on variables and values.

Python divides the operators into the following groups:

- Arithmetic operators
- Assignment operators
- Comparison operators
- Logical operators
- Identity operators
- Membership operators
- Bitwise operators



### 1. Arithmetic Operators
Used to perform basic mathematical operations.

| Operator | Description           | Example      |
|----------|-----------------------|--------------|
| `+`      | Adds two values       | `3 + 2 = 5`  |
| `-`      | Subtracts right from left | `5 - 2 = 3`  |
| `*`      | Multiplies values     | `4 * 2 = 8`  |
| `/`      | Divides left by right | `10 / 2 = 5.0` |
| `//`     | Floor division        | `7 // 2 = 3` |
| `%`      | Modulus (remainder)   | `7 % 2 = 1`  |
| `**`     | Exponentiation        | `2 ** 3 = 8` |

---

### 2. Comparison Operators
Used to compare two values and return a boolean (`True` or `False`).

| Operator | Description                   | Example         |
|----------|-------------------------------|-----------------|
| `==`     | Equal to                      | `3 == 3  True` |
| `!=`     | Not equal to                  | `3 != 2  True` |
| `>`      | Greater than                  | `5 > 2  True`  |
| `<`      | Less than                     | `2 < 5  True`  |
| `>=`     | Greater than or equal to      | `3 >= 3  True` |
| `<=`     | Less than or equal to         | `2 <= 3  True` |

---

### 3. Assignment Operators
Used to assign values to variables, optionally performing an operation.

| Operator | Description             | Example           |
|----------|-------------------------|-------------------|
| `=`      | Assigns a value         | `x = 5`           |
| `+=`     | Adds and assigns        | `x += 3`  `x = x + 3` |
| `-=`     | Subtracts and assigns   | `x -= 2`          |
| `*=`     | Multiplies and assigns  | `x *= 2`          |
| `/=`     | Divides and assigns     | `x /= 2`          |
| `//=`    | Floor divides and assigns | `x //= 2`       |
| `%=`     | Modulus and assigns     | `x %= 3`          |
| `**=`    | Power and assigns       | `x **= 2`         |

---

### 4. Logical Operators
Used to combine multiple conditions and return boolean values.

| Operator | Description                         | Example                  |
|----------|-------------------------------------|--------------------------|
| `and`    | True if both conditions are True    | `True and False  False` |
| `or`     | True if at least one is True        | `True or False  True`   |
| `not`    | Reverses the result                 | `not True  False`       |

---

### 5. Identity Operators
Used to check whether two variables point to the same object in memory.

| Operator | Description                          | Example              |
|----------|--------------------------------------|----------------------|
| `is`     | True if both refer to same object    | `x is y`             |
| `is not` | True if they do not refer to same object | `x is not y`     |

---

### 6. Membership Operators
Used to test if a value is present in a sequence (like list, string, tuple, etc).

| Operator | Description                     | Example                 |
|----------|----------------------------------|--------------------------|
| `in`     | True if value is present         | `'a' in 'cat'  True`    |
| `not in` | True if value is not present     | `'x' not in 'cat'  True`|

---

### 7. Bitwise Operators
Used to perform bit-level operations on integers.

| Operator | Description              | Example        |
|----------|--------------------------|----------------|
| `&`      | Bitwise AND              | `5 & 3 = 1`     |
| `|`      | Bitwise OR               | `5 | 3 = 7`     |
| `^`      | Bitwise XOR              | `5 ^ 3 = 6`     |
| `~`      | Bitwise NOT              | `~5 = -6`       |
| `<<`     | Left shift               | `5 << 1 = 10`   |
| `>>`     | Right shift              | `5 >> 1 = 2`    |

If-Else Statement
</h2>

</div>

In Python, an `if-else` statement is how you make your code perform different actions based on whether a certain condition is true or false. Think of it as a fork in the road for your program.

The basic idea is:

- If a condition is true, then do this.

- Otherwise (if it's false), then do that.

This allows your program to be smart and dynamic instead of just running the same code every single time.

---

### The Basic `if` Statement
The simplest form is a single `if` block, which runs only when its condition is `True`.

```python
temperature = 30

if temperature > 25:
    print("It's a hot day!")
```

---
### Adding an `else` Block
The `else` statement provides an alternative action that runs only when the `if` condition is `False`.

```python
age = 16

if age >= 18:
    print("You are old enough to vote.")
else:
    print("You are not old enough to vote yet.")
```

---
### Handling Multiple Conditions with `elif`
What if you have more than two possibilities? That's where `elif` (short for "else if") comes in. You can use as many `elif` blocks as you need to check for multiple different conditions in order.

```python
score = 85

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
else:
    print("Grade: F")
```

Python Loops
</h2>

</div>

n Python, a loop is a way to repeat a block of code multiple times. Instead of writing the same line of code over and over, you can use a loop to automate repetitive tasks. Think of it as putting a task on autopilot. 

Python has two main types of loops: `for` loops and `while` loops.

---

### `for` Loops: Iterating Through a Sequence
A for loop is used when you want to iterate over a sequence of items, like a list, a string, or a range of numbers. It runs once for each item in the sequence.

Imagine you have a to-do list and you need to check off each item one by one. A for loop does exactly that.



### Example 1: Looping through a list



```python
# A list of fruits
fruits = ["apple", "banana", "cherry"]

# Loop through each fruit in the list and print it
for fruit in fruits:
    print(fruit)
```

### Example 2: Looping through a string
A string is just a sequence of characters, so you can loop through it too.

```python
for letter in "Hello":
    print(letter)
```

---
### `while` Loops: Repeating as Long as a Condition is True
A `while` loop is used to repeat a block of code as long as a certain condition remains `True`.

Think of it like a child repeatedly asking "Are we there yet?" on a road trip. They keep asking as long as the condition (not being there yet) is true.

It's crucial to make sure the condition will eventually become `False`, otherwise, you'll create an infinite loop!

### Example: This loop will print numbers from 1 to 5.

```python
# Start counting from 1
count = 1

# Keep looping as long as count is less than or equal to 5
while count <= 5:
    print(f"The count is: {count}")
    count = count + 1 # Increment count, so the loop eventually ends

print("Loop finished!")
```

---
### Controlling Loops
Sometimes you need more control over how your loop behaves. Python provides two key statements for this:

- `break`: Immediately exits the current loop, regardless of the condition.

- `continue`: Skips the rest of the current iteration and jumps to the next one.

### `break` Example:
Stop the loop if a specific number is found.

```python
numbers = [1, 2, 3, 4, 5]
for num in numbers:
    if num == 4:
        print("Found the number 4! Stopping the loop.")
        break  # Exit the loop
    print(num)
```

### `continue` Example:
Print only the odd numbers by skipping the even ones.

```python
for i in range(1, 6): # Numbers from 1 to 5
    if i % 2 == 0:
        continue # Skip this iteration if the number is even
    print(i)
```
