<h2 style="text-align:center; font-family: 'Helvetica Neue', sans-serif; font-weight:bold; font-size: 26px;">
Lists
</h2>

</div>

- Lists are ordered, mutable (changeable) collections used to store items.

- Elements can be of any data type, and the list can be changed after creation (adding, removing, modifying values).

### How to Create a List

```python
# Creating an empty list
empty_list = []
print(f"Empty list: {empty_list}")

# Creating a list with elements
fruits = ["apple", "banana", "cherry"]
print(f"Fruits list: {fruits}")

# Creating a list with mixed data types
mixed_list = [1, "hello", 3.14, True]
print(f"Mixed list: {mixed_list}")
```

### List Operations

```python
# Accessing elements (indexing)
print(f"First fruit: {fruits[0]}")
print(f"Last fruit: {fruits[-1]}")

# Slicing lists
print(f"First two fruits: {fruits[0:2]}")

# Modifying elements
fruits[1] = "blueberry"
print(f"Modified fruits list: {fruits}")

# Adding elements
fruits.append("orange") # Add to the end
fruits.insert(1, "grape") # Insert at a specific index
print(f"Fruits after adding elements: {fruits}")

# Removing elements
fruits.remove("apple") # Remove by value
popped_fruit = fruits.pop() # Remove and return the last element
print(f"Fruits after removing elements: {fruits}")
print(f"Popped fruit: {popped_fruit}")

# List length
print(f"Number of fruits: {len(fruits)}")

# Checking if an element exists
print(f"Is 'cherry' in fruits? {'cherry' in fruits}")
```

Tuples
</h2>

</div>

- Tuples are ordered, immutable (cannot be changed) collections.

- Useful for representing fixed sets of valuese.g., coordinates, RGB values.

### How to Create a Tuple

```python
# Creating an empty tuple
empty_tuple = ()
print(f"Empty tuple: {empty_tuple}")

# Creating a tuple with elements
coordinates = (10, 20)
print(f"Coordinates tuple: {coordinates}")

# Creating a tuple with a single element (requires a trailing comma)
single_element_tuple = (5,)
print(f"Single element tuple: {single_element_tuple}")

# Creating a tuple without parentheses (tuple packing)
colors = "red", "green", "blue"
print(f"Colors tuple: {colors}")
```

### Tuple Example

```python
# Accessing elements
print(f"First coordinate: {coordinates[0]}")

# Tuples are immutable, so you cannot change elements
# coordinates[0] = 15 # This would raise a TypeError

# Tuple unpacking
x, y = coordinates
print(f"x: {x}, y: {y}")
```

<h2 style="text-align:center; font-family: 'Helvetica Neue', sans-serif; font-weight:bold; font-size: 26px;">
Dictionaries
</h2>

</div>

- Dictionaries store data as key-value pairs.

- Keys are unique and typically strings or numbers, while values can be any type.

### How to Create a Dictionary

```python
# Creating an empty dictionary
empty_dict = {}
print(f"Empty dictionary: {empty_dict}")

# Creating a dictionary with key-value pairs
person = {"name": "Alice", "age": 30, "city": "New York"}
print(f"Person dictionary: {person}")

# Creating a dictionary using the dict() constructor
another_person = dict(name="Bob", age=25, city="London")
print(f"Another person dictionary: {another_person}")
```

### Dictionary Operations

```python
# Accessing values
print(f"Person's name: {person['name']}")
print(f"Person's age (using .get()): {person.get('age')}") # Safer way to access, returns None if key not found

# Adding or modifying key-value pairs
person["job"] = "Engineer" # Add a new key-value pair
person["age"] = 31 # Modify an existing value
print(f"Modified person dictionary: {person}")

# Removing key-value pairs
del person["city"] # Remove by key
removed_value = person.pop("age") # Remove and return value for a key
print(f"Person after removing elements: {person}")
print(f"Removed age: {removed_value}")

# Getting keys, values, and items
print(f"Keys: {person.keys()}")
print(f"Values: {person.values()}")
print(f"Items: {person.items()}")

# Checking if a key exists
print(f"Does 'name' exist? {'name' in person}")
```

<h2 style="text-align:center; font-family: 'Helvetica Neue', sans-serif; font-weight:bold; font-size: 26px;">
Sets
</h2>

</div>

- Sets are unordered collections of unique elements.

- Useful for membership testing, removing duplicates, and set operations (union, intersection, etc.).

### How to Create a Set

```python
# Creating an empty set
empty_set = set()
print(f"Empty set: {empty_set}")

# Creating a set with elements (duplicates are removed)
numbers = {1, 2, 3, 3, 4, 5}
print(f"Numbers set: {numbers}")

# Creating a set from a list
my_list = [1, 2, 2, 3, 4, 4, 5]
set_from_list = set(my_list)
print(f"Set from list: {set_from_list}")
```

### Set Operations

```python
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

# Adding elements
set1.add(5)
print(f"Set1 after adding 5: {set1}")

# Removing elements
set1.remove(1) # Raises an error if element not found
set1.discard(10) # Does nothing if element not found
print(f"Set1 after removing 1 and discarding 10: {set1}")

# Union
union_set = set1.union(set2)
print(f"Union of set1 and set2: {union_set}")

# Intersection
intersection_set = set1.intersection(set2)
print(f"Intersection of set1 and set2: {intersection_set}")

# Difference
difference_set = set1.difference(set2)
print(f"Difference of set1 and set2: {difference_set}")

# Symmetric difference
symmetric_difference_set = set1.symmetric_difference(set2)
print(f"Symmetric difference of set1 and set2: {symmetric_difference_set}")

# Subset and superset
print(f"Is {{3, 4}} a subset of set1? {{3, 4}}.issubset(set1)")
print(f"Is set1 a superset of {{3, 4}}? set1.issuperset({{3, 4}})")
```

<h2 style="text-align:center; font-family: 'Helvetica Neue', sans-serif; font-weight:bold; font-size: 26px;">
Stacks and Queues
</h2>

</div>

Stacks and queues are linear data structures that follow specific rules for adding and removing elements.

- **Stack (LIFO - Last-In, First-Out):** The last element added is the first one to be removed. Think of a stack of plates.
- **Queue (FIFO - First-In, First-Out):** The first element added is the first one to be removed. Think of a queue of people waiting.

### Implementing a Stack (using a list)

```python
# Using a list as a stack
stack = []

# Push (add) elements
stack.append(1)
stack.append(2)
stack.append(3)
print(f"Stack after pushing: {stack}")

# Pop (remove) elements
popped_element = stack.pop()
print(f"Popped element: {popped_element}")
print(f"Stack after popping: {stack}")

popped_element = stack.pop()
print(f"Popped element: {popped_element}")
print(f"Stack after popping: {stack}")
```

### Implementing a Queue (using `collections.deque`)

While a list can be used as a queue, it's inefficient for large queues due to shifting elements when removing from the beginning. The `collections.deque` class is more efficient for this purpose.

```python
from collections import deque

# Using deque as a queue
queue = deque()

# Enqueue (add) elements
queue.append(1)
queue.append(2)
queue.append(3)
print(f"Queue after enqueuing: {queue}")

# Dequeue (remove) elements
dequeued_element = queue.popleft()
print(f"Dequeued element: {dequeued_element}")
print(f"Queue after dequeuing: {queue}")

dequeued_element = queue.popleft()
print(f"Dequeued element: {dequeued_element}")
print(f"Queue after dequeuing: {queue}")
```

Linked Lists
</h2>

</div>

- Linked List is a linear data structure where each item is a node containing data and a reference (link) to the next node.

- Singly Linked List: Each node links only to the next node.

### Node Class in a Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data  # Data stored in the node
        self.next = None  # Reference to the next node (initially None)
```

### Singly Linked List

```python
class SinglyLinkedList:
    def __init__(self):
        self.head = None  # The head of the list (initially None)

    # Method to print the linked list
    def print_list(self):
        current_node = self.head
        while current_node:
            print(current_node.data, end=" -> ")
            current_node = current_node.next
        print("None")
```

### Inserting a Node at the Beginning

```python
class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def print_list(self):
        current_node = self.head
        while current_node:
            print(current_node.data, end=" -> ")
            current_node = current_node.next
        print("None")

    def insert_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head  # The new node points to the current head
        self.head = new_node       # The head is updated to the new node

# Example usage
llist = SinglyLinkedList()
llist.insert_at_beginning(10)
llist.insert_at_beginning(20)
llist.insert_at_beginning(30)
print("Linked list after inserting at beginning:")
llist.print_list()
```

### Deleting a Node

```python
class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def print_list(self):
        current_node = self.head
        while current_node:
            print(current_node.data, end=" -> ")
            current_node = current_node.next
        print("None")

    def insert_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def delete_node(self, key):
        current_node = self.head

        # If the node to be deleted is the head
        if current_node and current_node.data == key:
            self.head = current_node.next
            current_node = None
            return

        # Search for the node to be deleted
        prev_node = None
        while current_node and current_node.data != key:
            prev_node = current_node
            current_node = current_node.next

        # If the key was not found
        if current_node is None:
            print(f"Node with key {key} not found.")
            return

        # Unlink the node from the linked list
        prev_node.next = current_node.next
        current_node = None

# Example usage
llist = SinglyLinkedList()
llist.insert_at_beginning(30)
llist.insert_at_beginning(20)
llist.insert_at_beginning(10)
print("Original linked list:")
llist.print_list()

llist.delete_node(20)
print("Linked list after deleting 20:")
llist.print_list()

llist.delete_node(50) # Trying to delete a non-existent node
print("Linked list after trying to delete 50:")
llist.print_list()
```

### Searching for a Node

```python
class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def print_list(self):
        current_node = self.head
        while current_node:
            print(current_node.data, end=" -> ")
            current_node = current_node.next
        print("None")

    def insert_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def search_node(self, key):
        current_node = self.head
        while current_node:
            if current_node.data == key:
                return True  # Node found
            current_node = current_node.next
        return False  # Node not found

# Example usage
llist = SinglyLinkedList()
llist.insert_at_beginning(30)
llist.insert_at_beginning(20)
llist.insert_at_beginning(10)
print("Linked list:")
llist.print_list()

print(f"Is 20 in the list? {llist.search_node(20)}")
print(f"Is 50 in the list? {llist.search_node(50)}")
```
